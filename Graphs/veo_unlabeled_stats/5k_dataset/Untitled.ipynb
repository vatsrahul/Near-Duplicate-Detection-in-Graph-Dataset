```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python

# Data for plotting Methods Vs Time
t = np.array(['Naive',
'+Loose',
'+Strict',
'+Static-\nIndexing\n+Partitioning',
'+Dynamic-\nIndexing\n+Partitioning'
          ])
s = np.array([53.963,
41.93,
41.733,
33.324,
37.265,
          ])
```

```python

fig, ax = plt.subplots()
ax.plot(t, s)

ax.set(xlabel='Methods', ylabel='time (sec)',
       title='Methods Vs Time')
ax.grid()

fig.savefig("MethodsVsTime.png")
plt.show()
```
