```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python

# Data for plotting
t = np.array(['Naive',
'+Loose',
'+Strict',
'+Static-\nIndexing\n+Partitioning',
'+Dynamic-\nIndexing\n+Partitioning'
          ])
s = np.array([8.622,
6.616,
6.662,
5.195,
5.526
          ])
```

```python

fig, ax = plt.subplots()
ax.plot(t, s)

ax.set(xlabel='Methods', ylabel='time (sec)',
       title='Methods Vs Time')
ax.grid()

fig.savefig("MethodsVsTime.png")
plt.show()
```
