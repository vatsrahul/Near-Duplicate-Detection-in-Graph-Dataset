```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python

# Data for plotting Methods Vs Time
t = np.array(['Naive',
'+Loose',
'+Strict',
'+Static-\nIndexing',
'+Dynamic-\nIndexing',
'+Partitioning'
          ])
s = np.array([8.622,
6.616,
6.662,
7.549,
6.733,
5.475
          ])
```

```python

fig, ax = plt.subplots()
ax.plot(t, s)

ax.set(xlabel='Methods', ylabel='time (sec)',
       title='Methods Vs Time')
ax.grid()

fig.savefig("MethodsVsTime.png")
plt.show()
```
