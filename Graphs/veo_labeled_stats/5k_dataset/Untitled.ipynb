```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python

# Data for plotting Methods Vs Time
t = np.array(['Naive',
'+Loose',
'+Strict',
'+Static-\nIndexing',
'+Dynamic-\nIndexing',
'+Partitioning'
          ])
s = np.array([166.961,
39.336,
39.195,
45.333,
40.793,
26.079
          ])
```

```python

fig, ax = plt.subplots()
ax.plot(t, s)

ax.set(xlabel='Methods', ylabel='time (sec)',
       title='Methods Vs Time')
ax.grid()

fig.savefig("MethodsVsTime.png")
plt.show()
```
