```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python

# Data for plotting Methods Vs Time
t = np.array(['Naive',
'+Loose',
'+Strict',
'+Static-\nIndexing',
'+Dynamic-\nIndexing',
'+Partitioning'
          ])
s = np.array([28.157,
6.156,
6.203,
7.149,
6.387,
4.196
          ])
```

```python

fig, ax = plt.subplots()
ax.plot(t, s)

ax.set(xlabel='Methods', ylabel='time (sec)',
       title='Methods Vs Time')
ax.grid()

fig.savefig("MethodsVsTime.png")
plt.show()
```
